```python
from tiled.client import from_uri
from load_dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

tiled_server_url = "http://localhost:8000"
tiled_api_key = TILED_KEY
```

```python
import numpy as np

def create_circle_image(height, width, center_x, center_y, radius):
    """Create a 2D image with a circle centered at (center_x, center_y)"""
    # Create coordinate grids
    y, x = np.meshgrid(np.arange(height), np.arange(width), indexing='ij')
    
    # Calculate distance from center
    distance = np.sqrt((x - center_x)**2 + (y - center_y)**2)
    
    # Create circle (1 inside, 0 outside)
    circle = (distance <= radius).astype(float)
    
    return circle
```

```python
sample_image = create_circle_image(100, 100, 50, 50, 10)

# Plot the sample image
import matplotlib.pyplot as plt
plt.imshow(sample_image, cmap='gray')
plt.title("Sample Circle Image")
plt.axis('off')
plt.show()
```

```python
# Connect to the Tiled server
tiled_client = from_uri(
    tiled_server_url,
    api_key=tiled_api_key
)

# Create the container path
container_path = "sample_test_now"
if container_path not in tiled_client:
    tiled_client.create_container(container_path)
tiled_client = tiled_client[container_path]

tiled_client.write_array(
        sample_image.astype(np.float32),
        metadata={
            "description": "Sample circle image",
            "motor1": "position_x"
        },
        key="data_smaller_now",
    )
```
